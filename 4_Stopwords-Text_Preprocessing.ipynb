```python
paragraph='''I stand before you not to ask for your sympathy, but to awaken your strength. A society cannot become great by keeping a section of its people in chains. A nation rises only when every citizen is treated with dignity, justice, and equality.
For centuries, many among us have been told that we are less worthy because of our birth. I reject this idea completely. A person's value is determined not by caste, but by character, knowledge, and hard work. No child enters this world with superiority or inferiority written upon them. These are divisions created by society, and what society creates, society can also change.
Remember this: freedom without equality is incomplete, and equality without education is impossible. Education is not merely a path to employment; it is the path to self-respect. A mind that is educated cannot be enslaved for long. Therefore, educate yourselves—not only to earn a livelihood, but to understand your rights and responsibilities.
I ask you not to depend on others for your progress. Stand on your own feet. Build your confidence. Organize yourselves peacefully and lawfully. Demand justice, but do so with reason, discipline, and courage. Social change achieved through knowledge is stronger and more lasting than change achieved through hatred.
Never believe that your destiny has already been decided. The future belongs to those who are willing to question injustice, challenge ignorance, and work tirelessly for a better society. Democracy is not just about casting a vote. Democracy lives when we respect one another, when the law protects every citizen equally, and when no person is denied opportunities because of birth.
I also say to those who enjoy privilege: equality does not weaken a nation—it strengthens it. Justice for one community is justice for the whole nation. The progress of India depends not on the success of a few, but on the dignity of all.
Young people, the future of this nation rests in your hands. Read more than you argue. Think more than you follow. Build more than you criticize. Let your knowledge become your greatest weapon and your character your greatest strength.
Let us build an India where no one is judged by caste, where every child has access to education, where women and men enjoy equal rights, where the Constitution is respected, and where liberty, equality, and fraternity are not merely words, but the foundation of our daily lives.
Let us move forward together with wisdom instead of prejudice, with courage instead of fear, and with unity instead of division.'''
```

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
stopwords.words("french")
```

```python
stopwords.words('german')
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
## apply stopwards and filter and then apply stemming 
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)# converting all the words into sentences 
```

```python
sentences
```

```python
from nltk.stem import SnowballStemmer
snowballstemmer=SnowballStemmer('english')
```

```python
## apply stopwards and filter and then apply snowball stemming 
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)# converting all the words into sentences 
```

```python
sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
## apply stopwards and filter and then apply lemmatizer
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word,pos='v') for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)# converting all the words into sentences 
```
